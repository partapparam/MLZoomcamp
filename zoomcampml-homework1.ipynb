```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df = pd.read_csv("/kaggle/input/housingdata/housing.csv", encoding="ISO-8859-1")
```

```python
print(len(df.columns))
pd.__version__
#2.0.3
```

```python
# Find columns with missing data
print(df.isnull().any(axis=0))
# total_bedrooms
```

```python
# find unique values in coulmns
print(df['ocean_proximity'].unique())
# 5
```

```python
# Find the mean of the median_house_value for homes that are Near Bay
print(df.loc[df['ocean_proximity'] == 'NEAR BAY', 'median_house_value' ].mean())
#259212
```

```python
#Calculate avg of total bedrooms
avg_total_bedrooms = df['total_bedrooms'].mean()
print(avg_total_bedrooms)
#fill the NaN/empty  values with the avg
updated_df = df['total_bedrooms'].fillna(avg_total_bedrooms)
#537.871
```

```python
# Check to see if the mean value changes
print(df['total_bedrooms'].isnull().any())
print(updated_df.isnull().any())
print(updated_df.mean())
#537.871
```

```python
island_df = df.loc[df['ocean_proximity'] == 'ISLAND']
print(island_df)
```

```python
updated_island_df = island_df.loc[:, ['housing_median_age', 'total_rooms', 'total_bedrooms']]
# Get the undelying numpy array
X = updated_island_df.values
print(X)
```

```python
# get the transpose of X - TX
xt = X.T
print(xt)
```

```python
# multiply the transpose of x and X
print(np.dot(xt, X))
print(xt @ X)
print(np.matmul(xt, X))

xtx = xt @ X
```

```python
# calculate the inverse of xtx
inv_xtx = np.linalg.inv(xtx)
print(inv_xtx)
```

```python
# create an array Y
y = [950, 1300, 800, 1000, 1300]
```

```python
w = (inv_xtx @ xt)
print(w)
w = w @ y
print(w)
```
